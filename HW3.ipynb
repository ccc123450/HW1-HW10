```
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x<0:
            return(1==0)
        else:
            s = str(x)
            l = len(s)
            x=0
            if l ==1:
                return(1==1)
            else:
                for i in range(l//2):
                    if s[i] != s[l-1-i]:
                        x=1
                        break     
        return(x==0)
```
