```
class Solution(object):
    def uniqueMorseRepresentations(self, words):
        """
        :type words: List[str]
        :rtype: int
        """
        result_letter = ""
        result_word = []
		
        morse_arr = [".-","-...","-.-.","-..",".","..-.","--.","....","..",".---","-.-",".-..","--","-.","---",".--.","--.-",".-.","...","-","..-","...-",".--","-..-","-.--","--.."]

        for word in words:
            for letter in word:
                result_letter = result_letter +morse_arr[ord(letter)-97]
            result_word.append(result_letter)
            result_letter = ""              
               
        return(len(set(result_word)))
```
