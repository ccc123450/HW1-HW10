```
class Solution(object):
    def numJewelsInStones(self, J, S):
        """
        :type J: str
        :type S: str
        :rtype: int
        """
        a=0
        for i in range(len(J)):
            for j in range(len(S)):
                if J[i]==S[j]:
                    a=a+1
        return a
```
