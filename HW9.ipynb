```
class Solution(object):
    def sortArrayByParity(self, A):
        """
        :type A: List[int]
        :rtype: List[int]
        """
        B=[]
        C=[]
        D=[]
        for i in range(len(A)):
            if A[i]%2==0:
                B.append(A[i])
            else:
                C.append(A[i])
        D=B+C
        return D
```
