```
class Solution(object):
    def defangIPaddr(self, address):
        """
        :type address: str
        :rtype: str
        """
        a=[]
        e=0
        for i in range(len(address)):
            if address[i]=='.':
                a.append(address[e:i])
                e=i+1
        a.append(address[e:])
        str4 = "[.]".join(a)
        return str4
```
