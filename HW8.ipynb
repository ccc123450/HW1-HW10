```
class Solution(object):
    def flipAndInvertImage(self, A):
        """
        :type A: List[List[int]]
        :rtype: List[List[int]]
        """
        a=A[0]
        for i in range(len(A)):
            for j in range(len(a)//2):
                b=A[i][j]
                c=A[i][len(a)-1-j]
                A[i][j]=c
                A[i][len(a)-1-j]=b
        for l in range(len(A)):
            for m in range(len(a)):
                if A[l][m]==1:
                    A[l][m]=0
                else:
                    A[l][m]=1
        return A
```
