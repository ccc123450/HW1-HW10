```
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        a=""
        if len(strs)==0: return a
    
        for i in xrange(len(min(strs))):
            c=strs[0][i]
            if all(b[i]==c for b in strs):
                a+=c
            else:
                break
        return a
```
